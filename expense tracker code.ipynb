```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the expenses data
df = pd.read_csv("expenses.csv")

# Total Spending Overview
total_spent = df['Amount'].sum()
max_expense = df.loc[df['Amount'].idxmax()]
min_expense = df.loc[df['Amount'].idxmin()]

print(f"\nðŸ’¸ Total Spent: â‚¹{total_spent}")
print(f"ðŸ”º Highest Expense: â‚¹{max_expense['Amount']} - {max_expense['Description']}")
print(f"ðŸ”» Lowest Expense: â‚¹{min_expense['Amount']} - {min_expense['Description']}")

# Category-wise Analysis
category_summary = df.groupby('Category')['Amount'].agg(['sum', 'count'])
category_summary['percentage'] = round((category_summary['sum'] / total_spent) * 100, 2)

print("\nðŸ“Š Category-wise Expense Summary:\n", category_summary)

# Optional Pie Chart
plt.figure(figsize=(7, 7))
plt.pie(category_summary['sum'], labels=category_summary.index, autopct='%1.1f%%')
plt.title("Expense Breakdown by Category")
plt.show()
```
